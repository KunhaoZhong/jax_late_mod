```python
import jax.numpy as jnp

bin=[0,0.5,1.0]

def smooth_piecewise(x, epsilon=0.1):
    # Transition function
    transition = lambda x: 0.5 * (jnp.tanh(x / epsilon) + 1)

    # Components of the piecewise function
    f1 = 5
    f2 = 2.0
    f3 = 4

    # Calculate blend weights
    w1 = transition(x + 1)  # Transition at x = -1
    w2 = 1 - transition(x - 1)  # Transition at x = 1

    # Combine the functions
    return f1 * w1 + f2 * (1 - w1) * w2 + f3 * (1 - w2)

# Test
import matplotlib.pyplot as plt
import numpy as np

x_vals = np.linspace(-2, 2, 400)
y_vals = [smooth_piecewise(x, epsilon=0.1) for x in x_vals]

plt.plot(x_vals, y_vals, label="Smooth Piecewise")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import jax.numpy as jnp

def smooth_custom_bins_piecewise(x, bins, values, epsilon=0.001):
    assert len(bins) == len(values) , "Number of bins should be 1 more than values"

    # Transition function
    transition = lambda x, edge: 0.5 * (jnp.tanh((x - edge) / epsilon) + 1)

    # Initial value
    result = values[0] * (1 - transition(x, bins[0]))

    # Iterating through the bins and adjusting the values accordingly
    for i in range(len(bins)-1):
        w1 = transition(x, bins[i])
        w2 = 1 - transition(x, bins[i+1])
        result += values[i] * w1 * w2

    # Final bin value
    result += values[-1] * transition(x, bins[-1])

    return result

# Test
import matplotlib.pyplot as plt
import numpy as np

x_vals = np.linspace(0, 2.5, 400)
bins = [0, 0.5, 1.0, 1.5]
values = [1, 2, 4, 2]
y_vals = [smooth_custom_bins_piecewise(x, bins, values, epsilon=0.05) for x in x_vals]

plt.plot(x_vals, y_vals, label="Custom Bins Piecewise")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def smooth_piecewise(x, bins, values, width=0.001):
    # Transition function
    transition = lambda x, edge: 0.5 * (np.tanh((x - edge) / width) + 1)

    # Initial value
    result = values[0] * (1 - transition(x, bins[0]))

    # Iterating through the bins and adjusting the values accordingly
    for i in range(len(bins)-1):
        w1 = transition(x, bins[i])
        w2 = 1 - transition(x, bins[i+1])
        result += values[i] * w1 * w2

    # Final bin value
    result += values[-1] * transition(x, bins[-1])

    return result

x_vals = np.linspace(0, 2.5, 400)
bins = [0, 0.5, 1.0, 1.5]
values = [1, 2, 4, 2]
y_vals = [smooth_piecewise(x, bins, values, width=0.05) for x in x_vals]

plt.plot(x_vals, y_vals, label="Custom Bins Piecewise")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```
